```python
df = pd.read_excel("MML datasheet 5.xlsx")
df.head()
```

```python
X=df.drop(labels=['Name','Team','Position','Age'],axis=1)
Y=df.drop(labels=['Name','Team','Position','Height(inches)','Weight(pounds)'],axis=1)
print(X)
print('Shape of X = ',X.shape)
print('Shape of Y = ',Y.shape)
```

```python
import math as math
import numpy as np
import matplotlib.pyplot as plt
X1=range(1,300,5)
Y1=[]
new=[73,195]
for k in range(1,300,5):
  K=k
  d=[]
  for i in range(500):
    a=[X.iat[i,0],X.iat[i,1]]
    d.append(math.dist(new,a))
  p=[]
  for i in range(500):
    a=Y.iat[i,0]
    p.append(a)
  l=[]
  q=[]
  for i in range(500):
    merged_list=(d[i],p[i])
    l.append(merged_list)
  l.sort()
  K1=l[:K]
  for i in range(len(K1)):
    q.append(K1[i][1])
  av=sum(q)/len(q)
  error=pow(av-24.11,2)
  Y1.append(error)
plt.plot(X1,Y1)
plt.xlabel('k values')
plt.ylabel('average error')
plt.show()
```
