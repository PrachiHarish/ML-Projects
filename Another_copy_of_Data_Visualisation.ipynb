```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Automobile_price_data__Raw_.csv")
print(data.info())
```

```python
data.head()
```

```python
data.drop(["symboling","normalized-losses"],axis=1,inplace=True)
```

```python
data=data.replace("?",np.nan)
```

```python
data.dropna(inplace=True)
```

```python
col=["bore","stroke","horsepower","peak-rpm","price"]
data[col]=data[col].apply(pd.to_numeric)
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,6))
plt.plot(data["price"].sort_values().unique())
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(8,6))
plt.title("Bivariate Plot Price vs Milage")
plt.plot(data.sort_values("price")["price"],data.sort_values("price")["city-mpg"],color="b")
```

Bar Plot

```python
plt.bar(x=data["make"].value_counts().index,height=data["make"].value_counts().values)
plt.xticks(rotation=90)
```

Histogram

```python
plt.hist(data["price"],bins=20)
```

Scatter plot

```python
sns.scatterplot(x=data["city-mpg"],y=data["price"],color="deeppink",marker="s",alpha=0.6)
```

Pie chart

```python
plt.figure(figsize = (7,7))
plt.pie(data['make'].value_counts(), labels = data['make'].value_counts().index, autopct = '%0.2f%%')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(x = data['city-mpg'], y = data['horsepower'], color = 'deeppink', marker = 's',alpha = 0.6,
label = 'horsepower vs city-mpg')
sns.scatterplot(x = data['city-mpg'], y = data['engine-size'], color = 'blue', alpha = 0.6,
label = 'engine-size vs city-mpg')
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.distplot(data['engine-size'], bins = 20, color = 'blue')
plt.show()
```
